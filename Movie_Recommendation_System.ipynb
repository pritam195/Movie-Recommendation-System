```python
import pandas as pandas
import numpy as np
```

```python
import csv
df = pandas.read_csv("/content/TMDB_movie_dataset_v11.csv", engine='python')
```

```python
df.head()
```

```python
df.info()
```

```python
df['tags'] = df['genres'] + df['keywords']
```

```python
df.head()
```

```python
new_df = df[['id', 'title', 'tags', 'vote_average', 'release_date']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
subset_size = 20000
vec_subset = vec[:subset_size]
sim = cosine_similarity(vec_subset)
```

```python
sim
```

```python
new_df[new_df['title'] == "Inception"].head()
```

```python
# Use the index within the subset
dist = sorted(list(enumerate(sim[0])), reverse=True, key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:10]:
  print(new_df.iloc[i[0]].title)
```

```python
def recommended(movie_name):
    movie_index_original = new_df[new_df['title'] == movie_name].index[0]

    if movie_index_original < subset_size:
        distance = sorted(list(enumerate(sim[movie_index_original])), reverse=True, key=lambda vec:vec[1])

        print(f"\nTop 10 recommendations similar to '{movie_name}':\n")
        for i in distance[1:11]:
            title = new_df.iloc[i[0]].title
            rating = new_df.iloc[i[0]].vote_average
            release = new_df.iloc[i[0]].release_date
            print(f"{title}  |  ⭐ IMDb Rating: {rating}  |  📅 Release: {release}")

```

```python
recommended("Iron Man")
```
